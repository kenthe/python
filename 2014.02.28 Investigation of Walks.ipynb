### Investigation of walks on MJO's test network

```python
import numpy as np
import networkx as nx
import matplotlib.pyplot as plt

k3   = np.array([[0, 1, 1],[1, 0, 1],[1, 1, 0]])
zero = np.zeros((3,3), dtype=int)
c0   = np.zeros((3,3), dtype=int)
c1   = np.zeros((3,3), dtype=int)
c2   = np.zeros((3,3), dtype=int)
c0[0,0] = 1; c0[0,2] = 1
c1[0,0] = 1; c1[0,1] = 1
c2[0,1] = 1; c2[0,2] = 1

h0 = np.hstack((k3, zero, zero, c0))
h1 = np.hstack((zero, k3, zero, c1))
h2 = np.hstack((zero, zero, k3, c2))
h3 = np.hstack((c0.T, c1.T, c2.T, zero))
```

```python
A = np.vstack((h0, h1, h2, h3))
G = nx.from_numpy_matrix(A)

nx.draw(G)
plt.show()
```

```python
A2  = A.dot(A)    # Finding matrix with 2-walks!
G2  = nx.from_numpy_matrix(A2)
G2r = nx.from_numpy_matrix(A*A2)  # restricted walks

A3  = A2.dot(A)   # Finding matrix of walks length 3
G3  = nx.from_numpy_matrix(A3)
G3r = nx.from_numpy_matrix(A*A3)
```

```python
network_layout=nx.spring_layout(G,iterations=300);

plt.subplot(221)
nx.draw(G,pos=network_layout, edge_color='k',width=3.0)
plt.title('G: One-walks')
plt.subplot(222)
nx.draw(G2,pos=network_layout,edge_color='r',width=3.0)
plt.title('G2: Walks of length 2')
plt.subplot(223)
nx.draw(G2r,pos=network_layout,edge_color='b',width=3.0)
plt.title('G2: Restricted walks of length 2')
plt.subplot(224)
nx.draw(G3r,pos=network_layout,edge_color='y',width=3.0)
plt.title('G3: Restricted walks of length 3')
```
